```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1=pd.read_csv('/content/mental-and-substance-use-as-share-of-disease -AI.csv')
df2=pd.read_csv('/content/prevalence-by-mental-and-substance-use-disorder _AI.csv')
```

```python
df=pd.merge(df2,df1)
```

```python
df.describe()
```

```python
df.set_axis(['Country','Year','Schizophrenia', 'Bipolar_disorder', 'Eating_disorder','Anxiety','drug_usage','depression','alcohol','mental_fitness'], axis='columns', inplace=True)
```

```python
df
```

```python
sns.heatmap(df.corr(),annot=True,cmap='Reds')
plt.plot()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
sns.kdeplot(data=df, x="mental_fitness", y="depression", cmap="Blues", fill=True)

plt.xlabel("Mental Fitness")
plt.ylabel("Depression")
plt.title("KDE Plot for Mental Fitness and Depression")

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i]=l.fit_transform(df[i])
```

```python
df.info()
```

```python
X = df.drop('mental_fitness',axis=1)
y = df['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
df.head()
# Linear Regression
# Regression Trees
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytrain, ytrain_pred)
print("The model performance for training set")
print("--------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)

print("Results for Random Forest Regressor")
print("--------------------------------------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

regression_tree = DecisionTreeRegressor(random_state=42)
regression_tree.fit(xtrain, ytrain)

y_pred = regression_tree.predict(xtest)

from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)
print("Results for Decision Tree Regressor")
print("Mean Squared Error:", mse)
print("R-squared:", r2)

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

multi_reg = LinearRegression()

multi_reg.fit(xtrain, ytrain)

y_pred = multi_reg.predict(xtest)

mse = mean_squared_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)
print("Results for Linear Regression")
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
